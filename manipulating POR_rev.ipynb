```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('POR.csv')
```

```python
df.head()
```

```python
df.columns = ['date', 'revenue']
```

```python
df['date'][0][-11:]
```

```python
df['date'] = df['date'].apply(lambda x: str(x)[-11:])
```

```python
df.drop(99, inplace = True)
```

```python
df['date'][0].split('-')[-1]
```

```python
def quarter(date):
    mon = date.split('-')[0]
    year = date.split('-')[-1]
    if mon == 'Mar':
        return 'Q1-'+ year
    elif mon == 'Jun':
        return 'Q2-'+ year
    elif mon == 'Sep':
        return 'Q3-'+ year
    else:
        return 'Q4-'+ year
    
```

```python
df['date'] = df['date'].apply(lambda x: quarter(x))
```

```python
df.index = df['date']
```

```python
df.drop('date', axis=1, inplace=True)
```

```python
df.to_csv('quarterly_revenue.csv',  encoding='utf-8')
```
